```python
import matplotlib.pyplot as plt
%matplotlib inline 
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("censusData", header = None,  names=['age','workclass','fnlwgt', 'education', 'education-num', ' marital-status', 'occupation','relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'yerlyincome'  ])
df
```

```python
df2 = df[:20]
df2
```

```python
# wykres słupkowy
p = df2['relationship'].value_counts()
p.plot.bar(figsize=(10,8), color ='maroon', width = 0.4)
```

```python
#wykres kołowy 
plot = df2['yerlyincome'].value_counts().plot.pie(figsize=(11, 6),shadow = True)
```

```python
#histogram 
hist = plt.hist(df2['age'], bins=7, edgecolor='black')
```

```python
#wykres rozrzutu 
plt.scatter(df2["age"], df2['relationship'])
```

```python
#wykres liniowy 
x = np.arange(0, 20, 2)
y = x**2

plt.plot(x,y)
plt.title('y=x**2')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
