4. Rozdziel kolumnę tshirt category na 3 kolumny: tshirt_type, która będzie mówiła o tym czy jest to koszulka, bluza czy coś innego, tshirt_size, która będzie informowała o rozmiarze koszulki, i tshirt_colour, która będzie informowała o kolorze. Skorzystaj z metody apply.

```python
import pandas as pd
import numpy as np

df = pd.read_csv("orders.csv")
df
```

```python
df['order_date'] = pd.to_datetime(df.order_date)
df.order_date.dt.strftime('%Y-%m-%d')
df
```

```python
df['tshirt_category'].unique()
```

```python
len(df['tshirt_category'].unique())
```

```python
df['tshirt_category']=df['tshirt_category'].apply(lambda x: x.replace('Wh', 'White'))
df['tshirt_category']=df['tshirt_category'].apply(lambda x: x.replace('Whiteite', 'White'))
df['tshirt_category']=df['tshirt_category'].apply(lambda x: x.replace('Bl', 'Black'))
df['tshirt_category']=df['tshirt_category'].apply(lambda x: x.replace('Blackack', 'Black'))
df['tshirt_category']=df['tshirt_category'].apply(lambda x: x.replace('Tshirt', 'T-shirt'))
df['tshirt_category']=df['tshirt_category'].apply(lambda x: x.replace('Shirt', 'shirt'))
df['tshirt_category']=df['tshirt_category'].apply(lambda x: x.replace('Tennis shirt', 'White T-shirt F'))
df['tshirt_category']=df['tshirt_category'].apply(lambda x: x.replace('Hoodie', 'White Hoodie F'))
```

```python
df['tshirt_category'].unique()
```

```python
df[['tshirt_colour', 'tshirt_type', 'tshirt_size']] = df['tshirt_category'].apply(lambda x: pd.Series(str(x).split(" ")))
df
```

```python
df.loc[(df['order_date'] > '2014-12-31') & (df['order_date'] <= '2016-08-2')] 
```
