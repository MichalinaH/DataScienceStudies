```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('FashionDataset.csv')
df
```

```python
df.shape
```

```python
df.info()
```

```python
len(df[df.duplicated()])
```

```python
df = df.drop_duplicates()
```

```python
df = df.rename(columns={'Deatils':'Details'})
```

```python
df.drop('Unnamed: 0', axis=1,  inplace=True)
df
```

```python
df = df.replace('Nan', np.NaN)
df
```

```python
df = df.dropna(subset=['Details'])
```

```python
df['Sizes']=df['Sizes'].str.replace('Size:',' ')
df
```

```python
df['Discount']=df['Discount'].str.replace('% off','')
```

```python
df['MRP']=df['MRP'].str.replace('Rs\n','')
```

```python
df['Discount']=df['Discount'].astype(float)
df['SellPrice']=df['SellPrice'].astype(int)
df['MRP']=df['MRP'].astype(float)
```

```python
df.info()
```

```python
df[['category', 'sex',]] = df['Category'].apply(lambda x: pd.Series(str(x).split("-")))
df
```

```python
df.drop('Category', axis=1,  inplace=True)
```

```python
count_brands = df["BrandName"].value_counts()
df["product count"] = df["BrandName"].apply(lambda x: count_brands[x] if not pd.isna(x) else np.nan) 
```

```python
plt.figure(figsize=(12,12))
plt.pie(df['category'].value_counts(), labels=df['category'].unique())
plt.show()
```

```python
plt.figure(figsize=(25,7))
plt.xticks(rotation=90)
sns.countplot(x=df['Discount'], data = df, order = df['Discount'].value_counts().index)
plt.show()
```
